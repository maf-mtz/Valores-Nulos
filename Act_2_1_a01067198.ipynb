```python
import pandas as pd
```

```python
df= pd.read_csv('Ventas_totales.csv')
df
```

```python
df.info()
```

```python
nulos=df.isnull().sum()
nulos
     
```

```python
#Copia del dataframe original
datos=df.copy()
```

### Columna 'salon_ventas'

```python
datos["salon_ventas"].max()
```

```python
datos["salon_ventas"].min()
```

```python
datos["salon_ventas"].mean()
```

```python
#Valores nulos de la columna 'salon_ventas'
#Sustituir valores nulos con promedio 
datos["salon_ventas"]=datos["salon_ventas"].fillna(round(df["salon_ventas"].mean(),1))
nulos=datos.isnull().sum()
nulos

#en la columna de 'salon_ventas había 6 datos nulos y lo sustitui con el promedio de los datos de la columna, 
#ya que son datos muy dispersos y cantidades grandes en millones
```

```python
valor_nulo=datos["salon_ventas"].isnull().sum()
valor_nulo
```

### Columna 'tarjetas_debito'

```python
datos["tarjetas_debito"].max()
```

```python
datos["tarjetas_debito"].min()
```

```python
#Esta columna tiene solo 1 datos nulo
import random
datos["tarjetas_debito"]= datos["tarjetas_debito"].fillna(random.randint(6438494, 55432191)) 
datos

#en este caso al ser solo un dato nulo, lo reemplacé con un dato aleatorio 
#que se encuentra en el rango de su minimo y maximo
```

```python
valor_nulo=datos["tarjetas_debito"].isnull().sum()
valor_nulo
```

### Columna 'tarjetas_credito'

```python
datos["tarjetas_credito"].mean()
```

```python
#en esta columna hay 7 datos nulos y los reemplazaré por el promedio de la columna 
#porque creo que es lo que afecta menos para el análisis de los datos,
#ya que son datos muy grandes y no se conoce en específico cuales serían los faltantes

datos["tarjetas_credito"]= datos["tarjetas_credito"].fillna(datos["tarjetas_credito"].mean()) 
datos
```

```python
valores_nulo=datos["tarjetas_credito"].isnull().sum()
valores_nulo
```

### Columna 'otros_medios'

```python
valores_nulo=datos["otros_medios"].isnull().sum()
valores_nulo

#Solo tenemos un datos nulo 
```

```python
#Sustituir valor nulos por valor no nulos hacia atrás backward fill" ("bfill")
datos["otros_medios"]= datos["otros_medios"].fillna(method="bfill") 
datos

#Me parece adecuado reemplazar ese dato con un dato no nulo que se encuentre antes que el nulo dentro de la columna
```

```python
valores_nulo=datos["otros_medios"].isnull().sum()
valores_nulo
#vemos que ya no hay datos nulos
```

### Columna 'subtotal_ventas_alimentos_bebidas'

### Columna 'panaderia'

```python
#En este caso al dato nulo que existe lo reemplazare 
#con un número que tambien se encuentra en la columna despues de este dato

#Sustituir valor nulos por valor no nulos hacia adelante  "forward fill" ("ffill")
datos["panaderia"]= datos["panaderia"].fillna(method="ffill") 
datos
```

```python
valores_nulos=datos["panaderia"].isnull().sum()
valores_nulos
```
